# Step1: Read the images from 3 folders

```python
import os as os
s=os.listdir('data/sunset')
g=os.listdir('data/green_nature')
w=os.listdir('data/water')
```

# Step 2 and 3: Convert images into pixel matrices &  Extract color distribution features from images

```python
from PIL import Image
data=[]
for i in s:
    im=Image.open('data/sunset/'+i)       
    p=im.load()
    wi, he = Image.open('data/sunset/'+i).size
    red=green=blue=[0] * 256 
    for i1 in range(wi):
        for j in range(he):
            red[p[i1,j][0]]=red[p[i1,j][0]]+1
            green[p[i1,j][1]]=green[p[i1,j][1]]+1
            blue[p[i1,j][2]]=blue[p[i1,j][2]]+1
    data.append([1]+red+green+blue)
for i in g:
    im=Image.open('data/green_nature/'+i)      
    p=im.load()
    wi, he = Image.open('data/green_nature/'+i).size
    red=green=blue=[0] * 256   
    for i1 in range(wi):
        for j in range(he):
            red[p[i1,j][0]]=red[p[i1,j][0]]+1
            green[p[i1,j][1]]=green[p[i1,j][1]]+1
            blue[p[i1,j][2]]=blue[p[i1,j][2]]+1
    data.append([2]+red+green+blue)
for i in w:
    im=Image.open('data/water/'+i)       
    p=im.load()
    wi, he = Image.open('data/water/'+i).size
    red=green=blue=[0] * 256 
    for i1 in range(wi):
        for j in range(he):
            red[p[i1,j][0]]=red[p[i1,j][0]]+1
            green[p[i1,j][1]]=green[p[i1,j][1]]+1
            blue[p[i1,j][2]]=blue[p[i1,j][2]]+1
    data.append([3]+red+green+blue)
data



```

# Step 4. Convert data into Pandas DataFrame

```python
import pandas as pd
dataframe = pd.DataFrame(data=data)
dataframe
```

# Step 5: Cluster the images

```python
from sklearn.cluster import KMeans
```

```python
no_of_clusters=3
km=KMeans(n_clusters=no_of_clusters)
Features=dataframe.drop(0,1)
km.fit(Features)
```

# Step 6: Evaluate clustering accuracy

```python
from sklearn.metrics import adjusted_rand_score as rand
```

```python
rand(km.labels_, dataframe[0])
```

#  Step7: Dimentionality Reduction

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_s=scaler.fit_transform(dataframe.drop(0,1))
```

# 768 dimensions

```python
from sklearn.decomposition import PCA
dimensions=768
P = PCA(n_components=dimensions)
P.fit(df_s)
dfs_red=P.transform(df_s)
P.explained_variance_ratio_.sum()
```

```python
no_of_clusters=3
km=KMeans(n_clusters=no_of_clusters)
F=pd.DataFrame(data=dfs_red)
km.fit(F)
rand(km.labels_, dataframe[0])
```
