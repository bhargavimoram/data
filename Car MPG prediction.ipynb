# Linear regression example - Predicting car MPG

# Step 1 : Load auto mobile MPG data


```python
import pandas as pd # data processing package; e.g. load csv files.. etc
```

```python
data = pd.read_csv('auto_mpg_data.csv')
```

# Step 2 : print top 10 rows of data

```python
data.head(10)
```

# Step 3 & 4: Create the Features Matrix and Labels vector

```python
Features = data
Features = Features.drop('model_name',1)
Features = Features.drop('mpg',1)
Labels = data['mpg']
Features.head(10) # notice dropped coloumns
```

# step 6: Normalize the features using standard scaler

```python
from sklearn.preprocessing import StandardScaler as ss
scaler = ss()
```

```python
scaler.fit(Features)
```

```python
Features = scaler.transform(Features);
```

# step 7: Split the data into training and test sets

```python
from sklearn import cross_validation as cv # The split the data into 80% training and 20% test.
Features_train, Features_test, Labels_train, Labels_test = cv.train_test_split(Features, Labels, test_size=0.2)
```

# Step 8:Traing a Linear Regression model using Stochastic Gradient Descent

```python
from sklearn import linear_model as lm
SGD_LR = lm.SGDRegressor()
SGD_LR.fit(Features_train, Labels_train)
```
